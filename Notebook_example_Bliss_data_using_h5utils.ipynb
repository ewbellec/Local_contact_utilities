# Patch several 2D kmaps

```python
##################################################################################################################################
###################################             Patch 2D SXDM maps            ####################################################
##################################################################################################################################
    
def patch2DKmaps(filename, index_scan_list, roi_name,
                 hexapod_signs = [+1, +1, +1], 
                 theta=0, # If there's an angle between the piezo and hexapod motors
                 fig_width=8):
    
    theta = np.deg2rad(theta)
    piezo_hexapod_dict = {'pix' : {'hexa_mot' :'thx', 'sign' : hexapod_signs[0]},
                          'piy' : {'hexa_mot' :'thy', 'sign' : hexapod_signs[1]},
                          'piz' : {'hexa_mot' :'thz', 'sign' : hexapod_signs[2]}}

    fig, ax = plt.subplots(figsize=(fig_width,fig_width))

    for index_scan in index_scan_list:
        sxdm = SXDM_Scan(filename, index_scan)

        mot1_hexa = sxdm.getMotorPosition(piezo_hexapod_dict[sxdm.motor1_name]['hexa_mot'])
        mot2_hexa = sxdm.getMotorPosition(piezo_hexapod_dict[sxdm.motor2_name]['hexa_mot'])
        
        mot1_piezo_rot = np.cos(theta)*sxdm.motor1 + np.sin(theta)*sxdm.motor2
        mot2_piezo_rot = np.cos(theta)*sxdm.motor2 - np.sin(theta)*sxdm.motor1
        
        mot1 = mot1_hexa + mot1_piezo_rot*piezo_hexapod_dict[sxdm.motor1_name]['sign']/1e3
        mot2 = mot2_hexa + mot2_piezo_rot*piezo_hexapod_dict[sxdm.motor2_name]['sign']/1e3
        
        roidata = sxdm.getRoiData(roi_name)

        # plot
        ax.pcolormesh(mot1, mot2, roidata, cmap='jet')
        ax.set_aspect('equal')

        ax.set_xlabel('{} [mm]'.format(piezo_hexapod_dict[sxdm.motor1_name]['hexa_mot']))
        ax.set_ylabel('{} [mm]'.format(piezo_hexapod_dict[sxdm.motor2_name]['hexa_mot']))
    fig.tight_layout()
```

```python
filename = '/data/visitor/ihmi1544/id01/20240301/RAW_DATA/Ni_b18s2p1/Ni_b18s2p1_BCDI_0020/Ni_b18s2p1_BCDI_0020.h5'
scanList = get_scans_title_str_hdf5(filename, '')
```

```python
index_scan_list = np.arange(18,29+1)
roi_name = 'mpx1x4_roi1'

patch2DKmaps(filename, index_scan_list, roi_name, 
             hexapod_signs = [+1, +1, +1], # [-1, +1, +1] for the new hexapod maybe
             theta=0, # If there's an angle between hexapod and piezo motors (in degrees)
             fig_width=8)
```
